```python
#test train split

from sklearn.model_selection import train_test_split

# dataset split
# creating a function for dataset split
def dataset(X,y):
    train_full_X, val_X, train_full_y, val_y = train_test_split(X, y,test_size=0.2,random_state = 0)

# Use the same function above for the validation set
    train_X, test_X, train_y, test_y = train_test_split(train_full_X, train_full_y, test_size=0.25,random_state = 0)
    return (train_X, val_X, train_y, val_y)
```

```python
from sklearn.preprocessing import MinMaxScaler

def scale(X):
    scaler = MinMaxScaler()
    scaler.fit(X)
    return X
```

```python
Model: Random forest classifier with top 6  features 
```

```python

from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier



#assigning the features under a list

features=['route','booking_origin','flight_duration','wants_extra_baggage', 'length_of_stay','num_passengers']
X= df[features]
#one hot encoding
X = pd.get_dummies(X, columns=features)
X= scale(X)
y= df.booking_complete       

X_train,X_val,y_train,y_val= dataset(X,y)

forest_model= RandomForestClassifier(random_state=1)
forest_model.fit(X_train, y_train)
preds= forest_model.predict(X_val)

print('ACCURACY: ',accuracy_score(y_val,preds)*100)
print('AUC score: ',roc_auc_score(y_val,preds))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.preprocessing import StandardScaler

# Assuming you have the dataset loading and model training code before this point
# ...

# Assigning the features under a list
features = ['route', 'booking_origin', 'flight_duration', 'wants_extra_baggage', 'length_of_stay', 'num_passengers']
X = df[features]

# One hot encoding
X = pd.get_dummies(X, columns=features)

# Scaling features using StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

y = df.booking_complete

X_train, X_val, y_train, y_val = dataset(X, y)

# Training Random Forest model
forest_model = RandomForestClassifier(random_state=1)
forest_model.fit(X_train, y_train)
rf_preds = forest_model.predict(X_val)

# Training XGBoost model
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
xgb_preds = xgb_model.predict(X_val)

# Calculating AUC scores
rf_auc_score = roc_auc_score(y_val, rf_preds)
xgb_auc_score = roc_auc_score(y_val, xgb_preds)

# Creating a bar chart with data labels
def plot_scores(rf_score, xgb_score, score_type1, score_type2):
    scores = pd.Series([rf_score, xgb_score], index=[score_type1, score_type2])
    scores = scores.sort_values(ascending=True)

    # Creating indices and ticks
    indices = np.arange(len(scores))
    ticks = list(scores.index)

    # Creating subplots
    fig, ax = plt.subplots(figsize=(8, 5))

    # Plotting scores
    bars = ax.barh(indices, scores, color=['blue', 'red'])

    # Adding labels and title
    ax.set_yticks(indices)
    ax.set_yticklabels(ticks)
    ax.set_xlabel('AUC Scores')
    ax.set_title(f"{score_type1} vs {score_type2}")

    # Adding data labels on bars
    for bar in bars:
        yval = bar.get_width()
        plt.text(yval, bar.get_y() + bar.get_height()/2, round(yval, 4), ha='left', va='center')

    # Displaying the plot
    plt.show()

# Calling the function to plot the bar chart with data labels
plot_scores(rf_auc_score, xgb_auc_score, 'Random Forest', 'XGBoost')
```
